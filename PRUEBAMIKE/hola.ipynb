```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Función de regresión no lineal a*x^b
def nonlinear_function(x, a, b):
    return a * np.power(x, b)

# Datos de ejemplo
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 8, 18, 32, 50])

# Ajuste de la curva a los datos
params, covariance = curve_fit(nonlinear_function, x_data, y_data)

# Parámetros óptimos
a_optimal, b_optimal = params

# Generar puntos predichos con la función ajustada
y_pred = nonlinear_function(x_data, a_optimal, b_optimal)

# Gráfico de los datos y la regresión no lineal
plt.scatter(x_data, y_data, label='Datos reales')
plt.plot(x_data, y_pred, label=f'Regresión no lineal: {a_optimal:.2f} * x^{b_optimal:.2f}', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.5, 2, 2.25, 2.4, 2.5])

# Parámetros estimados manualmente
a_estimado = 2.2
b_estimado = -1.6 / 2.2

# Función original
y_original = (a_estimado * x) / (x + b_estimado)

# Función linealizada
y_linealizada = a_estimado - (a_estimado * b_estimado) / (x + b_estimado)

# Graficar resultados
plt.scatter(x, y, label='Datos originales')
plt.plot(x, y_original, label='Función original', linestyle='--', marker='o')
plt.plot(x, y_linealizada, label='Función linealizada', linestyle='-', marker='x')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparación de y original y y linealizada')
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Definir la función no lineal
def modelo_no_lineal(x, a, b):
    return (a * x) / (x + b)

# Generar datos de ejemplo
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.5, 2, 2.25, 2.4, 2.5])

# Ajustar la regresión no lineal
parametros_optimos, covarianza = curve_fit(modelo_no_lineal, x, y)

# Imprimir los parámetros optimizados
a_optimo, b_optimo = parametros_optimos
print(f'a_optimo: {a_optimo}, b_optimo: {b_optimo}')

# Predecir valores con los parámetros optimizados
y_pred = modelo_no_lineal(x, a_optimo, b_optimo)

# Graficar los resultados
plt.scatter(x, y, label='Datos reales')
plt.plot(x, y_pred, label='Ajuste no lineal', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión no lineal: (a*x)/(x+b)')
plt.show()
```
