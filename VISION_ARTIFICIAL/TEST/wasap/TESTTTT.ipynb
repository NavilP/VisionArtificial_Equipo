```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagenes
fondo = plt.imread('../Fotos/fondo.jpg', 1)
sujeto = plt.imread('../Fotos/sujeto.jpg', 1)
paris = plt.imread('../Fotos/paris.jpeg', 1)

# Mostrar imagenes
plt.imshow(sujeto)
```

```python
plt.imshow(fondo)
```

```python
plt.imshow(paris)
```

```python
# Resta de imagenes
result = cv2.subtract(fondo, sujeto)
# Mostrar resta
plt.imshow(result)
```

```python
# Conversion a escala de grises
cv2.imwrite(
    "C:/Users/SZMik/OneDrive/Dokumente/Mike/VisualStudio/Fotos/gris.jpg", result)
imgGray = cv2.imread('../Fotos/gris.jpg', 0)
cv2.imwrite(
    "C:/Users/SZMik/OneDrive/Dokumente/Mike/VisualStudio/Fotos/gris2.jpg", imgGray)
imgGray2 = cv2.imread('../Fotos/gris2.jpg', 1)

# Mostrar escala de grises
plt.imshow(imgGray2)
```

```python
# Conversion a binaria
_, imgBin = cv2.threshold(imgGray2, 15, 255, cv2.THRESH_BINARY)
# Mostrar binaria
plt.imshow(imgBin)

```
